```python
def npToTensor(x):
    c,w,h = x.shape
    x = torch.from_numpy(x).to(device)
    x = torch.reshape(x, (c,1,w,h))
    return x.float()

def simplifyChannels(x):
    x = np.where(x == 3, 0, x) 
    x = np.where(x == 4, 3, x) 
    x = np.where(x == 5, 3, x) 
    x = np.where(x == 6, 4, x) 
    x = np.where(x == 7, 4, x) 
    x = np.where(x == 8, 5, x) 
    x = np.where(x == 9, 5, x) 
    x = np.where(x == 10 , 6, x) 
    x = np.where(x == 11, 7, x) 
    x = np.where(x == 12, 7, x)  
    x = np.where(x > 12, 0, x) 
    return (x / 9) + 0.1

def get_segmentation(x,scale_factor):
    classif = imp.loss.getTrainedModel()
    _,_,w,_ = x.shape
    if scale_factor > 0 :
        x = torch.nn.functional.interpolate(x, scale_factor=scale_factor)
            
    normalized = transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))(x)
    classifiedImage = classif(normalized)
    if scale_factor > 0 :
        classifiedImage = torch.nn.functional.avg_pool2d(classifiedImage, scale_factor)
            
    classifDisplay = imp.loss.generate_label(classifiedImage,w).to(device)
    classifPlain = imp.loss.generate_label_plain(classifiedImage,w)
    classifPlain = simplifyChannels(classifPlain)
    classifPlain = npToTensor(classifPlain)
    return classifPlain
```

```python
from torchvision.datasets.folder import ImageFolder
from torchvision import transforms
from torch.utils.data import DataLoader
import torch
import impaintingLib as imp
import numpy as np

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def test(folder="lfw",channels=4,seed=0,factorResize=4,batch_size=8):
    
    shuffle = seed == 0
    if not shuffle : 
        torch.manual_seed(seed)
        
    plot = imp.utils.Visu(gridSize=batch_size).plot_img
    resize = (120*factorResize, 120*factorResize)
    crop   = (64*factorResize, 64*factorResize)

    process = transforms.Compose([
         transforms.Resize(resize), 
         transforms.CenterCrop(crop),
         transforms.ToTensor()
    ])
    
    dataset = ImageFolder("./data/"+folder, process)
    data = DataLoader(dataset, num_workers=2, batch_size=batch_size, shuffle=shuffle)

    img,_ = next(iter(data))
    x = img.to(device)
    n,c,w,h = x.shape

    # Alter
    alter = imp.mask.Alter(resize=factorResize,seed=seed,test=True).irregularMask
    # x_prime = alter(imp.data.normalize(x))
    x_prime = alter(x)

    # Add Classifier
    classifPlain = get_segmentation(x,factorResize)
    
    # Predict
    model = imp.model.UNet(channels, netType="partial").to(device) 
    
    if channels == 4 :
        path = './modelSave/david/partial4channel_med'
        x_prime2 = torch.cat((x_prime,classifPlain),dim=1)
    else : 
        path = './modelSave/partial_22channels_highres.pth'
        x_prime2 = torch.cat((x_prime,classifiedImage),dim=1)
        
    model.load_state_dict(torch.load(path))
    x_hat = model(x_prime2)
    
    # Display
    plot(x)
    plot(classifPlain)
    #plot(classifiedImage[:,1:2])
    #plot(classifDisplay)
    #plot(imp.data.inv_normalize(x_prime))
    #plot(imp.data.inv_normalize(x_hat))
    
    plot(x_prime)
    plot(x_hat)
```
